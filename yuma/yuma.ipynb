```python
df = df.dropna(subset=['CustomerID'])
df = df.dropna(subset=['TransactionDate'])
```

```python
df['TotalAmount'] = df['TotalAmount'].abs()  # Convert negative values to positive
df['Quantity'] = df['Quantity'].abs()
df['TrustPointsUsed'] = df['TrustPointsUsed'].abs()
```

```python
df
```

```python
# Count rows with both PricePerUnit and TotalAmount as NaN
critical_missing_rows = df[df['PricePerUnit'].isna() & df['TotalAmount'].isna()]
critical_missing_count = len(critical_missing_rows)
print(f"Number of rows with both PricePerUnit and TotalAmount missing: {critical_missing_count}")
```

```python
percentage_missing = (critical_missing_count / len(df)) * 100
print(f"Percentage of data lost: {percentage_missing:.2f}%")
```

```python
# Plot distribution of PricePerUnit
plt.figure(figsize=(12, 6))
sns.histplot(df['PricePerUnit'].dropna(), bins=20, kde=True)
plt.title('Distribution of Price Per Unit')
plt.xlabel('Price Per Unit')
plt.ylabel('Frequency')
plt.show()

# Plot distribution of TotalAmount
plt.figure(figsize=(12, 6))
sns.histplot(df['TotalAmount'].dropna(), bins=20, kde=True)
plt.title('Distribution of Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()

#Plot for distribution of Discount
plt.figure(figsize=(12, 6))
sns.histplot(df['DiscountApplied'].dropna(), bins=20, kde=True)
plt.title('Distribution of DiscountApplied')
plt.xlabel('DiscountApplied')
plt.ylabel('Frequency')
plt.show()
```

```python
med_price = df['PricePerUnit'].median()
avg_price = df['PricePerUnit'].mean()
med_quan = df['Quantity'].median()
avg_quan = df['Quantity'].mean()
med_disc = df['DiscountApplied'].median()
avg_disc = df['DiscountApplied'].mean()
```

```python
print(med_price)
print(avg_price)
print(med_quan)
print(avg_quan)
print(med_disc)
print(avg_disc)
```

```python
df['PricePerUnit'].fillna(med_price, inplace=True)
df['Quantity'] = df['Quantity'].replace(0, med_quan)
df['DiscountApplied'].fillna(med_disc, inplace=True)
df['PaymentMethod'] = df['PaymentMethod'].fillna(df['PaymentMethod'].mode()[0])
```

```python
df
```

```python
df['TotalAmount'] = df.apply(
    lambda row: row['Quantity'] * row['PricePerUnit'] if pd.isna(row['TotalAmount']) or row['TotalAmount'] == 0 else row['TotalAmount'],
    axis=1
)
```

## **Data Preprocessing**

## **Q1. Steps Followed for Data Preprocessing**

    Handling Missing Values:

        Total Amount:
            Identified that missing values in the TotalAmount column could be imputed using the product of Quantity and PricePerUnit.
            Filled missing TotalAmount values by calculating Quantity * PricePerUnit, considering cases where Quantity or PricePerUnit were missing. In such cases, if both were missing, median imputation for Quantity and PricePerUnit was used to estimate the missing TotalAmount.

        Quantity:
            Analyzed the Quantity column and identified that it contained zero and negative values.
            Imputed missing Quantity values with the median of the non-zero, non-negative values in the column to ensure consistency and avoid skewed results due to outliers.

        Price Per Unit:
            For the PricePerUnit column, median imputation was chosen due to the high skew in the data, with the median value being 30.0 for the PricePerUnit.

        Discount Applied:
            Recognized that the DiscountApplied column might reflect discounts subtracted from the PricePerUnit. Due to lack of clarity on discount application, median imputation was used for missing values, with the median discount being 10.

        Payment Method:
            Imputed missing values in the PaymentMethod column using the mode, which represents the most frequently occurring value. This ensures consistency and reflects the most common payment method used in the dataset.

        Transaction Date:
            Removed rows with missing TransactionDate as this column is crucial for time-based analysis and its absence could not be reliably imputed.

    Handling Negative Values:
        Identified that negative values in Quantity, PricePerUnit, and TrustPointsUsed might not be valid for a sales dataset.
        Decided to convert all negative values to positive values to ensure consistency and avoid errors in analysis.

    Column Drop Decision:
        Evaluated the DiscountApplied column and decided to drop it if its explanation was unclear and it could not be reliably used for imputation or analysis.

    Data Validation:
        After preprocessing, validated the dataset to ensure that all missing values were handled appropriately and no data anomalies remained.

### **Summary**

These preprocessing steps ensured that the dataset was cleaned, consistent, and ready for further analysis or modeling. By addressing missing values through appropriate imputation methods, handling negative values, and validating data integrity, we prepared the dataset to be more reliable and useful for deriving insights.

        Sum: To calculate the total revenue generated from each product, customer, or transaction period. This is crucial for financial analysis and revenue tracking.

    DiscountApplied

    TransactionID
        Count: To count the number of transactions, which provides a measure of overall transaction volume.

    CustomerID
        Count: To determine the number of unique customers, which helps in understanding customer base size.
        Sum/Average of Purchase Metrics: To calculate total or average purchases per customer, giving insight into customer spending behavior.

    ProductID
        Count: To determine the number of transactions involving each product, which shows product popularity.
        Sum/Average of Purchase Metrics: To calculate total or average sales metrics (e.g., quantity, total amount) for each product, providing insight into product performance.

    PaymentMethod
        Count: To determine the frequency of each payment method, helping to understand payment preferences.

### Calculate the Pearson correlation coefficient between the variables. A coefficient close to +1 or -1 indicates a strong linear relationship.

```python
correlation = df[['DiscountApplied', 'TotalAmount']].corr(method='pearson')
print(correlation)
```

### Fit a simple linear regression model to the data. If the R-squared value is high (close to 1), it indicates a good linear fit.

```python
import statsmodels.api as sm

X = df['DiscountApplied']
y = df['TotalAmount']
X = sm.add_constant(X)  # Adds a constant term to the predictor
model = sm.OLS(y, X).fit()
print(model.summary())
```
