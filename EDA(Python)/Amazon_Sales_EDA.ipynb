```python
#plt.figure(figsize = (10,5))
sns.set(rc = {'figure.figsize' : (10,5)})
ax = sns.countplot(x = 'Size', data = df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
qty = df.groupby(['Size'], as_index = False)['Quantity'].sum().sort_values(by = 'Quantity', ascending = False)

ax = sns.barplot(x = 'Size', y = 'Quantity', data = qty)

for i in ax.containers:
    ax.bar_label(i)
```

*Among all sizes, M reigns supreme in the popularity contest, followed by L and XL, according to the sales data showcased in the graph.*

```python
plt.figure(figsize = (30,10))
ax = sns.countplot(x = 'ship-state', data = df)
plt.xlabel('Ship State')
plt.ylabel('Shipments')
plt.xticks(rotation = 90)
for i in ax.containers:
    ax.bar_label(i)
```

```python
top_10 = df['ship-state'].value_counts().head(10)
ax = sns.countplot(data = df[df['ship-state'].isin(top_10.index)], x = 'ship-state')
plt.xlabel('Top 10 Shipped States')
plt.ylabel('Shipments')
plt.xticks(rotation = 90)
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_state = df.groupby(['ship-state'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(data = sales_state, x = 'ship-state',y= 'Amount')
```

```python
df.groupby(['ship-state'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
```

```python
sales_city = df.groupby(['ship-city'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(data = sales_city, x = 'ship-city',y= 'Amount')
```

```python
df.groupby(['ship-city'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
```

*Bangalore claims the city revenue crown, while Maharashtra secures the statewide throne.*

```python
sales_service = df.groupby(['ship-service-level'], as_index=False)['B2B'].sum().sort_values(by='B2B', ascending=False)

sns.set(rc={'figure.figsize':(8,5)})
ax = sns.barplot(data = sales_service, x = 'ship-service-level',y= 'B2B')

for i in ax.containers:
    ax.bar_label(i)
```

*B2B buyers crave speed: expedited shipping nearly doubles standard options.*

```python
shirt_orders = df[df['Category'] == "T-shirt"]
state_counts = shirt_orders.groupby('ship-state').size().sort_values(ascending=False).head(10)
ax = sns.barplot(x=state_counts.index, y=state_counts.values)
sns.set(rc = {'figure.figsize' : (10,5)})
plt.xlabel("States")
plt.ylabel("Number of Shirt Orders")
plt.title("States with the Most Shirt Orders")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
for i in ax.containers:
    ax.bar_label(i)
```

*Maharashtra reigns supreme in the land of T-shirts!*

```python
shirt_orders = df[df['Category'] == "T-shirt"]
city_counts = shirt_orders.groupby('ship-city').size().sort_values(ascending=False).head(10)
ax = sns.barplot(x=city_counts.index, y=city_counts.values)
sns.set(rc = {'figure.figsize' : (15,5)})
plt.xlabel("Cities")
plt.ylabel("Number of T-Shirt Orders")
plt.title("States with the Most T-Shirt Orders")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
for i in ax.containers:
    ax.bar_label(i)
```

*Bangalore is the reigning champion of T-shirt orders!*

```python
import pandas as pd
import seaborn as sns

# Assuming your DataFrame is named `df`, and it has columns 'ship-postal-code' and 'Status'

cancelled_orders = df[df['Status'] == "Cancelled"]
pincode_counts = cancelled_orders.groupby('ship-postal-code').size()

# Sort by the number of cancelled orders directly
top_pincodes = pincode_counts.sort_values(ascending=False).head(10)


# Create the barplot using sorted values
ax = sns.barplot(x=top_pincodes.index, y=top_pincodes.values, order=top_pincodes.index)

# Add labels in the correct order
for i in ax.containers:
    ax.bar_label(i)
```
